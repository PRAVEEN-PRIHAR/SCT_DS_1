```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student_score.csv")
print(df.head())
```

```python
df.isnull().sum()
```

# Drop unnamed column

```python
df = df.drop("Unnamed: 0",axis = 1)
print(df.head())
```

# Gender Distribution

```python
sns.countplot(data =df, x = 'Gender')
plt.title("Gender Distribution")
plt.figure(figsize = (4,3))

plt.show()
```

# from the above chart i anlaysis that : the number of females data is more than number of males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb,annot = True)
plt.title("Relationship between Parent Education and Student Score")
plt.figure(figsize = (4,4))
plt.show()
```

# from above chart i conclued that the education of parents that have good impact on student scores.

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent Marital Status and Student Score")
plt.figure(figsize = (4,4))
plt.show()
```

# from above chart i have concluded that there is no impact on students' scores due to their parents marital status
